```python
import pickle
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import colors as colors
import seaborn as sns
# sns.set_style("whitegrid")
from matplotlib import rc
rc('text', usetex=True)

def get_traj(data):
    res = data['traj']
    res = sorted(res,key = lambda t:t['iter'])
    iter_s = [t['iter'] for t in res]
    distr_s = [t['distri'] for t in res]
    batch_size = data['batch_size']

    return (iter_s,distr_s,batch_size)
```

# Gradient Descent Convergence

```python
filename = '../experiments/fashionmnist_gd_batchsize_1200/loss_distribution.pkl'
with open(filename,'rb') as f:
    data = pickle.load(f)
iter_s, distr_s, batch_size = get_traj(data)
distr_s = np.asarray(distr_s)

loss_s=distr_s.mean(axis=1)
plt.plot(iter_s,loss_s)
```

```python
plt_idx = [10,50,90,120,190,220]

plt.figure(figsize=(16,8))
for i,idx in enumerate(plt_idx,1):
    plt.subplot(2,3,i)
    sns.distplot(distr_s[idx],kde=False,color='black')
    plt.title('Iteration: %d'%(iter_s[idx]),fontsize=15)
plt.savefig('figs/fashionmnist_gd_loss_distribution.pdf',bbox_inches='tight')
```

```python
filename = '../experiments/fashionmnist_sgd_batchsize_20/loss_distribution.pkl'
with open(filename,'rb') as f:
    data = pickle.load(f)
iter_s, distr_s, batch_size = get_traj(data)
distr_s = np.asarray(distr_s)

loss_s=distr_s.mean(axis=1)
plt.plot(iter_s,loss_s)
```

## From Gradient Descent To Stochastic Gradient Descent

```python
filename = '../experiments/fashionmnist_gd2sgd_zoomin/loss_distribution.pkl'
with open(filename,'rb') as f:
    data = pickle.load(f)
iter_s, distr_s, batch_size = get_traj(data)
distr_s = np.asarray(distr_s)

loss_s=distr_s.mean(axis=1)

plt.figure(figsize=(10,7))
plt.tick_params(axis='both', which='major', labelsize=15)
plt.plot(iter_s,loss_s,'o-',lw=4,markersize=15)
plt.xlim([3040,3070])
plt.ylabel('Training Loss',fontsize=20)
plt.xlabel('Number of Iterations',fontsize=20)
plt.savefig('figs/fashionmnist_switch_transition.pdf',bbox_inches='tight')
```
